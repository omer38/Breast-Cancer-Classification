```python
def train_val_test_split(df, y_column, test_size=0.15, val_size=0.10, seed=38):
    np.random.seed(seed)
    shuffled_indices = np.random.permutation(df.index)
    
    test_set_size = int(df.shape[0] * test_size)
    val_set_size = int(df.shape[0] * val_size)

    test_indices = shuffled_indices[:test_set_size]
    val_indices = shuffled_indices[test_set_size:test_set_size + val_set_size]
    train_indices = shuffled_indices[test_set_size + val_set_size:]

    train_df = df.iloc[train_indices]
    val_df = df.iloc[val_indices]
    test_df = df.iloc[test_indices]

    X_train = train_df.drop(columns=[y_column])
    y_train = train_df[y_column]
    X_val = val_df.drop(columns=[y_column])
    y_val = val_df[y_column]
    X_test = test_df.drop(columns=[y_column])
    y_test = test_df[y_column]

    return X_train, X_val, X_test, y_train, y_val, y_test
```

```python
def standardize_data(X_train,X_test):
    mu = X_train.mean()
    sigma = X_train.std()
    
    X_train_s = (X_train - mu) / sigma
    X_test_s = (X_test - mu) / sigma
    
    return X_train_s,X_test_s
```

```python
def metrics(y_pred,y_test):
    tp, fp, tn, fn = 0, 0, 0, 0
    for i in range(len(y_test)):
        true, pred = y_test.iloc[i], y_pred[i]
        if true == 1 and pred == 1:
            tp += 1
        elif true == 0 and pred == 1:
            fp += 1
        elif true == 0 and pred == 0:
            tn += 1
        elif true == 1 and pred == 0:
            fn += 1
    if (tp + fp + tn + fn) != 0:
        accuracy = (tp + tn) / (tp + fp + tn + fn)
    else:
        accuracy = 0

    if (tp + fp) != 0:
        precision = tp / (tp + fp)
    else:
        precision = 0

    if (tp + fn) != 0:
        recall = tp / (tp + fn)
    else:
        recall = 0

    if (precision + recall) != 0:
        f1 = 2 * (precision * recall) / (precision + recall)
    else:
        f1 = 0

    conf_dict = {'true=1':[tp,fn],'true=0':[fp,tn]}
    conf_matrix = pd.DataFrame(data=conf_dict)
    conf_matrix.index = ['pred=1','pred=0']

    return accuracy,precision,recall,f1,conf_matrix
```

# PCA

Steps for PCA:
We normalized pixel before doing PCA.
1) Center the data
2) Compute the covarience matrix (X.T . X)/N
3) Do eigenvalue decomposition to find eigenvalue and eigenvectors(use numpy library)
4) Order eigenvalues in descending order, the eigenvector corresponds to highest eigenvalue will show the direction which captures most varience.
5) Return sorted eigenvalues and sorted eigenvectors(Principal components) in the same index

```python
def PCA(X_train):
    X_centered = X_train - np.mean(X_train,axis = 0)# STEP 1-calculate mean of each feature and subtract
    n_rows = X_centered.shape[0]
    cov_mat = (X_centered.T@X_centered)/n_rows # STEP2
    eig_val, eig_vec = np.linalg.eig(cov_mat) # STEP 3
    i = np.argsort(eig_val)[::-1] #STEP 4
    eig_val_sorted = eig_val[i]
    eig_vec_sorted = eig_vec[:,i]
    return eig_val_sorted,eig_vec_sorted
```

```python
def reconstruction(images,eig_vec,k):
    mean_val = np.mean(images,axis = 0)
    cent_images = images - mean_val
    z = np.dot(cent_images,eig_vec[:, :k]) # project features to k dimension new shape(60k,k)
    new_images = np.dot(z, eig_vec[:, :k].T) #(60k,k)*(k,784) = (60k,784) which is original image
    new_images = new_images + mean_val
    return new_images
```

# MODEL1: LOGISTIC REGRESSION

```python
class LogisticRegression:
    def __init__(self,l_rate = 0.01, iterations = 1000):
        self.l_rate = l_rate
        self.iterations = iterations
        self.w = None 

    def logistic_function(self,x):
        # P(Y=1|X=Xk)
        return 1/(1+np.exp(-x))

    def fit_logistic(self,X,y):
        m,n = X.shape #m=row, n=column/feature
        self.w = np.zeros(n)

        #Do gradient descent algortihm to update weights
        for i in range(self.iterations):
            z = np.dot(X,self.w)
            pred = self.logistic_function(z)
            
            error = y - pred
            #use cross entropy loss to calculate gradient.
            gradient = np.dot(X.T,error) / m #Sum of errors weighted by its features
            self.w = self.w + self.l_rate*gradient

    def predict_prob(self,X):
        z = np.dot(X,self.w)
        return self.logistic_function(z)
    
    def predict(self,X,threshold = 0.5):
        probs = self.predict_prob(X)
        predictions = []
        for prob in probs:
            if prob >= threshold:
                predictions.append(1)
            else:
                predictions.append(0)
    
        return np.array(predictions)
```

```python
def kfold_cv(df, k=5, l_rate=0.01, iterations=1000):
    fold_size = len(df) // k
    accuracies = []
    precisions = []
    recalls = []
    f1_scores = []

    for fold in range(k):
        start = fold * fold_size
        if fold < k - 1:
            end = start + fold_size
        else:
            end = len(df)
        test_df = df.iloc[start:end]
        train_df = pd.concat([df.iloc[:start], df.iloc[end:]])

        X_train = train_df.drop('diagnosis', axis=1)
        y_train = train_df['diagnosis']
        X_test = test_df.drop('diagnosis', axis=1)
        y_test = test_df['diagnosis']

        X_train_s, X_test_s = standardize_data(X_train, X_test)

        logistic_model = LogisticRegression(l_rate=l_rate, iterations=iterations)
        logistic_model.fit_logistic(X_train_s, y_train)

        y_pred = logistic_model.predict(X_test_s)

        accuracy,precision,recall,f1,conf_matrix = metrics(y_pred,y_test)

        accuracies.append(accuracy)
        precisions.append(precision)
        recalls.append(recall)
        f1_scores.append(f1)

    print("Average Accuracy:", np.mean(accuracies))
    print("Average Precision:", np.mean(precisions))
    print("Average Recall:", np.mean(recalls))
    print("Average F1-Score:", np.mean(f1_scores))

kfold_cv(df)
```

```python
class MLP:
    def __init__(self, input_size=30, hidden_size=64,rand_seed=41):
        np.random.seed(rand_seed)
        self.w1 = np.random.randn(input_size, hidden_size)
        self.b1 = np.zeros(hidden_size)
        self.w2 = np.random.randn(hidden_size, 1)
        self.b2 = np.zeros(1)

    def sigmoid(self, x):
        return 1 / (1 + np.exp(-x))

    def sigmoid_derivative(self, x):
        return x * (1 - x)

    def forward(self, x):
        self.z1 = np.dot(x, self.w1) + self.b1
        self.a1 = self.sigmoid(self.z1)
        self.z2 = np.dot(self.a1, self.w2) + self.b2
        self.a2 = self.sigmoid(self.z2)
        return self.a2.squeeze()

    def backward(self, x, y, y_pred, learning_rate):
        
        y = np.array(y).reshape(-1, 1)
        y_pred = y_pred.reshape(-1, 1)
        
        error1 = y_pred - y
        sig_deriv = self.sigmoid_derivative(y_pred)

        d_w2 = np.dot(self.a1.T, error1 * sig_deriv)
        d_b2 = np.sum(error1 * sig_deriv, axis=0)

        error2 = np.dot(error1 * sig_deriv, self.w2.T) * self.sigmoid_derivative(self.a1)
        d_w1 = np.dot(x.T, error2)
        d_b1 = np.sum(error2, axis=0)

        self.w1 -= learning_rate * d_w1
        self.b1 -= learning_rate * d_b1
        self.w2 -= learning_rate * d_w2
        self.b2 -= learning_rate * d_b2

    def train(self, X_train, y_train, epochs, learning_rate):
        for epoch in range(epochs):
            y_pred = self.forward(X_train)
            self.backward(X_train, y_train, y_pred, learning_rate)

    def evaluate(self, X_test, y_test, threshold):
        y_pred = self.forward(X_test) > threshold
        accuracy, precision, recall, f1, conf_matrix = metrics(y_pred, y_test)
        return accuracy, precision, recall, f1, conf_matrix

model_mlp = MLP(input_size = 30, hidden_size=64)
model_mlp.train(X_train_s, y_train, epochs=100, learning_rate=0.01)
accuracy,precision,recall,f1,conf_matrix = model_mlp.evaluate(X_test_s, y_test,0.5)
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1-Score:", f1)
conf_matrix
```

```python
import numpy as np

class Node:
    def __init__(self, feature=None, threshold=None, left=None, right=None, value=None):
        self.feature = feature
        self.threshold = threshold
        self.left = left
        self.right = right
        self.value = value

class DecisionTree:
    def __init__(self, max_depth=None):
        self.max_depth = max_depth
        self.root = None

    def entropy(self, y):
        _, counts = np.unique(y, return_counts=True) # first placeholder will hold unique values and the second one holds count
        p = counts / counts.sum()
        entropy_sum = 0
        for prob in p:
            if prob > 0:
                entropy_sum -= prob * np.log2(prob)
        return entropy_sum

    def split(self, X, y, feature, threshold):
        left_indices = []
        right_indices = []

        # Iterate through each sample in the dataset
        for i in range(len(X)):
            # Check if the feature value of the sample is less than or equal to the threshold
            if X[i, feature] <= threshold:
                left_indices.append(i)
            else:
                # If the feature value is greater than the threshold
                right_indices.append(i)

        # Use the indices to select the corresponding target values
        left_y = y[left_indices]
        right_y = y[right_indices]

        return left_y, right_y

    
    def information_gain(self, y, left, right):
        ent_parent = self.entropy(y)
        ent_left = self.entropy(left)
        ent_right = self.entropy(right)
        w_left = len(left) / len(y)
        w_right = len(right) / len(y)
        ig = ent_parent - w_left * ent_left - w_right * ent_right
        return ig
    
    def best_split(self, X, y):
        # Initialize variables to store the best feature and threshold found so far
        best_feature_index = None
        best_threshold = None
        best_information_gain = -1

        # Iterate over all features in the dataset
        for feature_index in range(X.shape[1]):
            # Find the unique values for this feature
            unique_values = np.unique(X[:, feature_index])

            # Iterate over all unique values of the feature
            for value in unique_values:
                # Split the dataset based on the current feature and value
                left_split, right_split = self.split(X, y, feature_index, value)

                # Skip the split if it results in an empty group
                if len(left_split) == 0 or len(right_split) == 0:
                    continue

                # Calculate the information gain from this split
                information_gain = self.information_gain(y, left_split, right_split)

                # Update the best split if this one is better
                if information_gain > best_information_gain:
                    best_information_gain = information_gain
                    best_feature_index = feature_index
                    best_threshold = value

        # Return the best feature index and threshold found
        return best_feature_index, best_threshold

    def build_tree(self, X, y, depth=0):
        # Check if all target values are the same or if the maximum depth has been reached
        if len(np.unique(y)) == 1 or depth == self.max_depth:
            # Return a leaf node with the value of the target variable
            return Node(value=np.unique(y)[0])

        # Find the best feature and threshold to split on
        best_feature, best_threshold = self.best_split(X, y)

        # If no split can be made, return a leaf node with the most common target value
        if best_feature is None:
            return Node(value=np.bincount(y).argmax())

        # Initialize empty lists to hold indices for the left and right splits
        left_indices = []
        right_indices = []

        # Iterate through each sample in the dataset
        for i in range(len(X)):
            # Check if the value of the best feature in this sample is less than or equal to the best threshold
            if X[i, best_feature] <= best_threshold:
                left_indices.append(i)  # Add the index to the left_indices list
            else:
                right_indices.append(i) # Add the index to the right_indices list

        # Use the lists of indices to create the left and right subsets of X and y
        left_subtree = self.build_tree(X[left_indices], y[left_indices], depth + 1)
        right_subtree = self.build_tree(X[right_indices], y[right_indices], depth + 1)


        # Return the current node storing the best feature, threshold, and its children
        return Node(feature=best_feature, threshold=best_threshold, left=left_subtree, right=right_subtree)


    def fit(self, X, y):
        self.root = self.build_tree(X, y)

    def predict_sample(self, x, node):
        if node.value is not None:
            return node.value
        if x[node.feature] <= node.threshold:
            return self.predict_sample(x, node.left)
        else:
            return self.predict_sample(x, node.right)

    def predict(self, X):
        # Initialize an empty list to store the predictions
        predictions = []

        # Iterate through each sample in the dataset
        for sample in X:
            # Predict the class for the current sample and add it to the predictions list
            predicted_class = self.predict_sample(sample, self.root)
            predictions.append(predicted_class)

        # Convert the list of predictions to a NumPy array and return it
        return np.array(predictions)
    
```

# Model 4 - Random Forest

```python
class RandomForest:
    def __init__(self, n_trees=100, max_features=None, max_depth=None):
        self.n_trees = n_trees
        self.max_features = max_features
        self.max_depth = max_depth
        self.trees = []

    def bootstrap_sample(self, X, y):
        """
        Create a bootstrap sample of the training data.
        Bootstrap sampling is a method to randomly select data with replacement,
        which means the same data point can be selected more than once.
        
        Parameters:
        X: array-like, shape = [n_samples, n_features]
           Training input samples where n_samples is the number of samples
           and n_features is the number of features.

        y: array-like, shape = [n_samples]
           The target values (class labels) of the training samples.

        Returns:
        X_sample: array-like, shape = [n_samples, n_features]
                  The bootstrapped training input samples.

        y_sample: array-like, shape = [n_samples]
                  The target values (class labels) of the bootstrapped samples.
        """
        # Generate random indices for selecting data with replacement
        indices = np.random.randint(0, len(X), len(X))

        # Use the generated indices to create the bootstrap sample
        X_sample = X[indices]
        y_sample = y[indices]

        return X_sample, y_sample

    def feature_subset(self, X):
        """
        Select a random subset of features from the dataset.

        This method is used in Random Forest to provide each tree with a random subset
        of features. This enhances the diversity of the model and reduces overfitting.

        Parameters:
        X: array-like, shape = [n_samples, n_features]
           Training input samples where n_samples is the number of samples
           and n_features is the number of features.

        Returns:
        (X_subset, selected_features): tuple
            X_subset: array-like, shape = [n_samples, n_selected_features]
                      The subset of training input samples with selected features.

            selected_features: array-like
                               Indices of the selected features. Useful for tracking
                               which features are used by each tree.
        """

        # Check if 'max_features' is set and less than the total number of features
        if self.max_features is not None and self.max_features < X.shape[1]:
            # Randomly select 'max_features' number of feature indices
            selected_features = np.random.choice(X.shape[1], self.max_features, replace=False)

            # Extract the columns corresponding to the selected features
            X_subset = X[:, selected_features]

            # Return the subset of features and the indices of these features
            return X_subset, selected_features

        else:
            # If 'max_features' is not set or is greater than the total number of features,
            # use all features without any selection
            selected_features = None
            X_subset = X

            # Return the original dataset and None for the selected features
            return X_subset, selected_features

    def fit(self, X, y):
        self.trees = []
        for _ in range(self.n_trees):
            X_sample, y_sample = self.bootstrap_sample(X, y)
            X_sample, features = self.feature_subset(X_sample)
            tree = DecisionTree(max_depth=self.max_depth)
            tree.fit(X_sample, y_sample)
            self.trees.append((tree, features))

    def majority_vote(self, predictions):
        """
        Perform a majority vote on the predictions from all trees.

        This function calculates the average prediction from all trees and
        rounds it to the nearest integer. In a classification context, this
        is equivalent to a majority vote among the predictions.

        Parameters:
        predictions: array-like
                     The sum of predictions from all trees for each sample.

        Returns:
        majority_votes: array-like
                        The final predicted class for each sample after
                        performing the majority vote.
        """
        # Calculate the average prediction and round to the nearest integer
        majority_votes = np.round(predictions / self.n_trees).astype(int)
        return majority_votes

    def predict(self, X):
        """
        Predict class for X.

        The predicted class of an input sample is a vote by the trees in the forest,
        weighted by their probability estimates of the class.

        Parameters:
        X: array-like, shape = [n_samples, n_features]
           Input samples to be predicted.

        Returns:
        predictions: array-like, shape = [n_samples]
                     The predicted classes for each input sample.
        """
        predictions = np.zeros(len(X))
        for tree, features in self.trees:
            if features is not None:
                # Subset the features using the selected feature indices
                preds = tree.predict(X[:, features])
            else:
                preds = tree.predict(X)
            predictions += preds

        # Use the majority vote function to determine the final prediction
        return self.majority_vote(predictions)

```
